```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib notebook 
```

```python
LSH = pd.read_excel('Results.xlsx', sheet_name='LHS')
PF_LSH = pd.read_excel('Results.xlsx', sheet_name='PF_LHS')
RDN = pd.read_excel('Results.xlsx', sheet_name='RDN')
PF_RDN = pd.read_excel('Results.xlsx', sheet_name='PF_RDN')
```

```python
LSH_t=pd.DataFrame.to_numpy(LSH)
PF_LSH_t=pd.DataFrame.to_numpy(PF_LSH)
RDN_t=pd.DataFrame.to_numpy(RDN)
PF_RDN_t=pd.DataFrame.to_numpy(PF_RDN)
```

```python
# def NormalizeData(data):
#     return (data - np.min(data)) / (np.max(data) - np.min(data))

# LSH_n=NormalizeData(LSH_t)
# PF_LSH_n=NormalizeData(PF_LSH_t)
# RDN_n=NormalizeData(RDN_t)
# PF_RDN_n=NormalizeData(PF_RDN_t)
```

```python
y=LSH_t[:,0]
x=np.array([1, 2, 3, 4, 5])

LHS_loss=np.around(LSH_t[:,[1,2,3,4,5]], decimals=3)
PF_LSH_=PF_LSH_t[:,[1,2,3,4,5]]

RDN_loss=np.around(RDN_t[:,[1,2,3,4,5]], decimals=3)
PF_RDN_=PF_RDN_t[:,[1,2,3,4,5]]
```

```python
fig, (ax1, ax2) = plt.subplots(1,2,figsize=(8,4.5))

# cmap_reversed1 = plt.cm.get_cmap('Spectral_r')
cmap_reversed1 = plt.cm.get_cmap('Wistia')#, vmin=0, vmax=10)
cmap_reversed2 = plt.cm.get_cmap('Wistia')
plt.rcParams.update({'font.family':'Times New Roman'})
size = 12

im1 = ax1.imshow(LHS_loss, cmap = cmap_reversed1, vmin=0, vmax=0.5)
im2 = ax2.imshow(PF_LSH_, cmap = cmap_reversed2, vmin=1, vmax=10)

train = ['1', '2', '3', '4', '5']
degree = ['10', '20', '30', '40', '50']

# # Show all ticks and label them with the respective list entries

ax1.set_xticks(np.arange(len(train)), labels=train)
ax1.set_yticks(np.arange(len(degree)), labels=degree)

ax2.set_xticks(np.arange(len(train)), labels=train)
ax2.set_yticks(np.arange(len(degree)), labels=degree)

# Loop over data dimensions and create text annotations.
for i in range(len(x)):
    for j in range(len(y)):
        text = ax1.text(j, i, LHS_loss[i, j], ha="center", va="center", color="k", size=size)

        
        
ax1.set_title('LSE with LHS', size=size)
ax1.set_xlabel('Polynomial RS degree', size=size)
ax1.set_ylabel('Training batch sizes', size=size)


ax1.tick_params(axis='x', labelsize=size)
ax1.tick_params(axis='y', labelsize=size)
        
# plt.xlabel('Polyonomial RS degree', size=size)
# plt.ylabel('Training batches', size=size)
# plt.title('LSE with LHS', size=size)

for i in range(len(x)):
    for j in range(len(y)):
        text = ax2.text(j, i, PF_LSH_[i, j], ha="center", va="center", color="k", size=size)

ax2.set_title('Failure probability with LHS', size=size)
ax2.set_xlabel('Polynomial RS degree', size=size)
# ax2.set_ylabel('Training batches', size=size)

ax2.tick_params(axis='x', labelsize=size)
ax2.tick_params(axis='y', labelsize=size)
plt.savefig("LSE_LHS.svg")

fig.tight_layout()
plt.show()
```
